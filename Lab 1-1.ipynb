```python
import numpy as np
# Needed for plotting
import numpy.matlib
import matplotlib.pyplot as plt
```

```python
#Initialisation of variables
a=3.21
b=5.12
noise = np.random.normal(0,1,150)
# 0 is the mean of the normal distribution you are choosing from
# 1 is the standard deviation of the normal distribution
# 150 is the number of elements you get in array noise
X = np.random.uniform(0,1,150)
Y= a * X + b + noise
plt.scatter(X, Y, color = "orange", s = 20)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
a = b = 0
step = 0.15
epoch = 2000
k= len(X)
for i in range(epoch): 
        Y_predicted = a*X + b
        pa = (-2/k) * sum(X * (Y - Y_predicted))
        pb = (-2/k) * sum(Y - Y_predicted)
        a = a - step * pa
        b = b - step * pb
print (a, b)
Y_predicted = a*X + b
plt.scatter(X, Y,color = "orange",s=20) 
plt.plot([min(X), max(X)], [min(Y_predicted), max(Y_predicted)], color='green')
plt.show()
```
